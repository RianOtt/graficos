```python
plt.figure(figsize=(9, 6)) # Define o tamanho da figura para o gráfico
top_10_brands = df['brand'].value_counts().head(10) # Conta a ocorrência de cada marca e seleciona as 10 maiores
plt.bar(top_10_brands.index, top_10_brands.values, color='purple', edgecolor='black') # Cria um gráfico de barras com as 10 marcas mais frequentes
plt.title('Contagem dos Top 6 Fabricantes de Veículos', fontsize=16) # Adiciona um título ao gráfico
plt.xlabel('Fabricante', fontsize=12) # Define o rótulo do eixo X
plt.ylabel('Número de Veículos', fontsize=12) # Define o rótulo do eixo Y
plt.xticks(rotation=45, ha='right') # Rotaciona os rótulos do eixo X
plt.tight_layout() # Ajusta o layout para evitar sobreposição de elementos
```

```python
plt.figure(figsize=(9, 4)) # Define o tamanho da figura para o gráfico
plt.plot(avg_price_by_year['year_of_reference'], avg_price_by_year['avg_price_brl'], marker='o') # Cria um gráfico de linha com o ano de referência no eixo X e o preço médio no eixo Y, adicionando marcadores aos pontos
plt.title('Evolução do Preço Médio dos Veículos por Ano de Referência', fontsize=16) # Adiciona um título ao gráfico
plt.xlabel('Ano de Referência', fontsize=12) # Define o rótulo do eixo X
plt.ylabel('Preço Médio (BRL)', fontsize=12) # Define o rótulo do eixo Y
plt.grid(True) # Adiciona uma grade ao gráfico para facilitar a leitura
plt.tight_layout() # Ajusta o layout para evitar sobreposição de elementos
plt.show() # Exibe o gráfico
```
